```python
#Imputing Misiing Values
df_processed = df

#Categorical Columns
cat_cols_fill_none = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu',
                     'GarageCond', 'GarageQual', 'GarageFinish', 'GarageType',
                     'BsmtFinType2', 'BsmtExposure', 'BsmtFinType1', 'BsmtQual', 'BsmtCond',
                     'MasVnrType']

#Replacing the missing values in the catgorical columns with None
for cat in cat_cols_fill_none:
    df_processed[cat] = df_processed[cat].fillna('None')
    
#Grouping by neighbourhood and filling the missing values of LotFrontage on the basis ofthe median LotForntage of the neighbourhood
df_processed['LotFrontage'] = df_processed.groupby('Neighborhood')['LotFrontage'].transform( lambda x:x.fillna(x.median()))

# Garage: GarageYrBlt, GarageArea and GarageCars these are numerical columns, replace with zero
for col in ['GarageYrBlt', 'GarageArea', 'GarageCars']:
    df_processed[col] = df_processed[col].fillna(int(0))
    
# MasVnrArea replace with zero
df_processed['MasVnrArea'] = df_processed['MasVnrArea'].fillna(int(0))

#Using the mode value to fiil in the missing values in the Electrical column
df_processed['Electrical'] = df_processed['Electrical'].fillna(df_processed['Electrical'].mode())[0]

# There is no use of the Utilities column so dropping it from the dataframe
df_processed = df_processed.drop(['Utilities'] , axis=1)

# Getting the count to assure that there are no missing values in the dataframe
df_processed.isnull().apply(sum).max()
```

```python
# Removing the noisy data
num_attributes = df_processed.select_dtypes(exclude='object')
high_quant = df_processed.quantile(.999)

for i in num_attributes.columns:
    df_processed = df_processed.drop(df_processed[i][df_processed[i]>high_quant[i]].index)
df_processed.info()
```

```python
###Remove highly correlated features
#Remove the attributes identified for removing while viewing scatterplots and corr values
attributes_drop = ['MiscVal', 'MoSold', 'YrSold', 'BsmtFinSF2','BsmtHalfBath','MSSubClass',
                   'GarageArea', 'GarageYrBlt', '3SsnPorch']
df_processed = df_processed.drop(attributes_drop , axis=1)
```

```python
#Transforming categorical attributes
from sklearn.preprocessing import OneHotEncoder
cat_encoder = OneHotEncoder()
df_processed_1hot = cat_encoder.fit_transform(df_processed)
df_processed_1hot
```

```python
# Importing the pipeline modules
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

#Separating the features and the target variable
df_X = df_processed.drop('SalePrice' ,axis=1)
df_Y = df['SalePrice'].copy()

# Get the list of numerical and categorical attributes separately
num_attribs = list(df_X.select_dtypes(exclude='object'))
cat_attribs = list(df_X.select_dtypes(include='object'))

# Numerical Pipeline to impute any missing values with median and scale attributes
num_pipeline = Pipeline([('imputer',SimpleImputer(strategy= 'median')) , ('std_scaler', StandardScaler())])
```

```python
# Building a single transformer to transform both numerical and categorical data
full_pipeline = ColumnTransformer([('num',num_pipeline,num_attribs),('cat',OneHotEncoder(),cat_attribs)])

#Description before applying transforms
print(df_Y.describe())

#Applying log transformation to SalePrice
df_Y_prepared =np.log(df_Y)

#Run the transformation pipeline on all other attributes
df_X_prepared = full_pipeline.fit_transform(df_X)

#Description before applying transforms
print(df_Y_prepared.describe())

df_X_prepared
```
